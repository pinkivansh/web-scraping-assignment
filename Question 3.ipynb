```python
from bs4 import BeautifulSoup
import requests
import pandas as pd


url="https://m.imdb.com/list/ls056092300/"
page=requests.get(url)
soup=BeautifulSoup(page.content,"html.parser")
#print(soup.prettify())
movie_name=[]
year=[]
rating=[]

for i in soup.find_all('span', class_="h4"):
    movie_name.append(i.text)
    

for j in soup.find_all('span', class_="nowrap"):
    year.append(j.text)
    
for k in soup.find_all('span', class_="imdb-rating"):
    rating.append(k.text)
#rating
year 
#movie_name
```
