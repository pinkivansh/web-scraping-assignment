```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
url="https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating"
page=requests.get(url)
soup=BeautifulSoup(page.content,"html.parser")
#print(soup.prettify())
movie_name=[]
year=[]
rating=[]

movie_data=soup.findAll('div', attrs={'class':'lister-item mode-advanced'})

for store in movie_data:
    name= store.h3.a.text
    movie_name.append(name)
    year_of_relese=store.h3.find('span', class_='lister-item-year text-muted unbold').text.replace('(', '').replace(')', '')
    year.append(year_of_relese)
    
    rate= store.find('div', class_='inline-block ratings-imdb-rating').text.replace('\n', '')
    rating.append(rate)
    
df=pd.DataFrame({'Movie Name': movie_name, 'year of Release' : year, 'Ratings': rating})
df
```
